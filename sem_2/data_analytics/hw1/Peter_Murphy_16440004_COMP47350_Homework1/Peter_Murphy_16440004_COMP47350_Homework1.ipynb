```python
# Add 'race' and 'ethnicity' columns
df['race'] = ""
df['ethnicity'] = ""

for index, row in df.iterrows():
    # Change current_status to Confirmed/Not confirmed
    if (row['current_status'] == "Laboratory-confirmed case"):
        df.iloc[index]['current_status'] = 'Confirmed'
    else:
        df.iloc[index]['current_status'] = 'Not confirmed'
        
    # Change 'Missing' to None for 'sex' and 'age_group'
    if (row['sex'] == "Missing"):
        df.iloc[index]['sex'] = None
    if (row['age_group'] == "Missing"):
        df.iloc[index]['age_group'] = None
    else:
        df.iloc[index]['age_group'] = int(df.iloc[index]['age_group'][0:2])+5
        
    # Split race_ethnicity into 2 columns
    if (df.iloc[index]['race_ethnicity_combined'] == "Missing"):
        df.iloc[index]['race'] = None
        df.iloc[index]['ethnicity'] = None
    elif (df.iloc[index]['race_ethnicity_combined'] == "Unknown"):
        df.iloc[index]['race'] = 'Unknown'
        df.iloc[index]['ethnicity'] = 'Unknown'
    elif(df.iloc[index]['race_ethnicity_combined'] == "Hispanic/Latino"):
        df.iloc[index]['race'] = "Unknown"
        df.iloc[index]['ethnicity'] = df.iloc[index]['race_ethnicity_combined'].split("/")[0]
    else:
        df.iloc[index]['race'] = df.iloc[index]['race_ethnicity_combined'].split(", ")[0]
        df.iloc[index]['ethnicity'] = df.iloc[index]['race_ethnicity_combined'].split(", ")[1]
        
    # Change 'Missing' to None for 'hosp_yn', 'icu_yn' and 'medcond_yn'
    if (df.iloc[index]['hosp_yn'] == "Missing"):
        df.iloc[index]['hosp_yn'] = None
    if (df.iloc[index]['icu_yn'] == "Missing"):
        df.iloc[index]['icu_yn'] = None
    if (df.iloc[index]['medcond_yn'] == "Missing"):
        df.iloc[index]['medcond_yn'] = None
        
# Remove race_ethnicity column
df = df.drop(['race_ethnicity_combined'], axis = 1)
```

```python
# Save cleaned dataframe as cleaned_1.csv
df.to_csv('cleaned_v1.csv', index = False)
```

```python
# Part (4).
# Add 'time_to_report' feature by getting the difference between the earliest date and report date
df['time_to_report'] = df['cdc_report_dt'] - df['cdc_case_earliest_dt']
df['time_to_report'].describe()
```

```python
# day_1 = 1 day (a datetime64 time delta variable)
day_1 = df['time_to_report'][0]*1/15

# Define a function to return the percentage of death cases for x 'days to report'
def death_perc(x):
    day_n = x*day_1
    y = df.loc[(df['time_to_report'] >= day_n) & 
               (df['death_yn'] == "Yes")].shape[0]*100/df.loc[(df['time_to_report'] >= day_n)].shape[0]
    return y

# Create an array of days and an array of corresponding death percentages
weeks = 10
days = np.linspace(0,7*weeks,7*weeks+1)
y = np.empty(len(days))
for i in range(len(days)):
    y[i] = death_perc(days[i])

# Plot death percentage vs number of days
fig = plt.figure()
fig.set_figwidth(1.5 * fig.get_figwidth())
fig.set_figheight(1.5 * fig.get_figheight())
plt.plot(days,y)
plt.scatter(days,y, s=10, color='k')
plt.title('Death % Vs Time to Report')
plt.ylabel('Percentage of deaths')
plt.xlabel('Days')
plt.grid(True)
plt.savefig('time_to_report.png')
```
