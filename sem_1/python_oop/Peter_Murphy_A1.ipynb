```python
# Define person class with ID, name, balance, active in event
class person:
    def __init__(self, ID, name):
        self.ID = ID
        self.name = name
        self.balance = 0
        self.active = False

# Define event class, with list of members, list of transactions, total cost, cost per person
class event:
    def __init__(self):
        self.members = []
        self.transactions = []
        self.total = 0
        self.each = 0
    
    # Method to add person to event
    def add_person(self, person):
        if person.active == False:
            self.members.append(person)
            person.active = True
        else:
            print(person.name, "is already active in an event.")
    
    # Method to add transaction to event
    def add_transaction(self, name, cost, payer_ID):
        # Create list of ID's in event
        ID_list = []
        for i in range(len(self.members)):
            ID_list.append(self.members[i].ID)
            
        # Check that name is a string, cost is a positive number and payer_ID is in the list of ID's
        if isinstance(name, str) and ((isinstance(cost, int) or isinstance(cost, float)) and cost > 0) and payer_ID in ID_list:
            self.transactions.append([name, cost, payer_ID])  
            # Increase the total cost of the event by the cost of the transaction
            self.total += cost
            self.each += cost/len(self.members)
            for i in range(len(self.members)):
                # Decrease each member's balance by their share of the cost 
                self.members[i].balance -= cost/len(self.members)
                if self.members[i].ID == payer_ID:
                    # Increase the payer's balance by the cost 
                    self.members[i].balance += cost
        else:
            print("Invalid Transaction")
            
    # Method to reconcile
    def reconcile(self):
        print('Total: £{}, that is £{:.2f} each.\n'.format(self.total, self.each))
        
        # Sort members by balance
        def get_balance(person):
            return person.balance
        self.members.sort(key=get_balance, reverse = True)
        
        # Print each balance
        for i in range(len(self.members)):
            print('{} has balance £{:.2f}'.format(self.members[i].name, self.members[i].balance))
        print('')        
        
        # Calulate who pays who
        for i in range(len(self.members)):
            for j in range(len(self.members)):
                if int(self.members[i].balance) != 0 and self.members[-(j+1)].balance != 0:
                    if abs(self.members[i].balance) >= abs(self.members[-(j+1)].balance):
                        print('{} pays {} £{:.2f}'.format(self.members[-(j+1)].name, self.members[i].name, abs(self.members[-(j+1)].balance)))
                        self.members[i].balance += self.members[-(j+1)].balance
                        self.members[-(j+1)].balance = 0
                    else:
                        print('{} pays {} £{:.2f}'.format(self.members[-(j+1)].name, self.members[i].name, abs(self.members[i].balance)))
                        self.members[-(j+1)].balance += self.members[i].balance
                        self.members[i].balance = 0
                        
        
```

```python
# Sample 1
p1 = person(1, "Annie")
p2 = person(2, "Sally")
p3 = person(3, "Bill")

e1 = event()

e1.add_person(p1)
e1.add_person(p2)
e1.add_person(p3)

e1.add_transaction("Tickets", 180, 1)
e1.add_transaction("Dinner", 75, 2)
e1.add_transaction("Drinks", 19, 3)
e1.add_transaction("Taxi", 16, 3)

e1.reconcile()
```
